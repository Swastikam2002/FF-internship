**Import python libraries**

```python
import numpy as np
import pandas as pd
```

**Read Economic Data csv file**

```python
# Define the columns to exclude
columns_to_exclude = ['Unnamed: 0', 'Country', 'RICS', 'TIMESTAMP']

# Read the CSV file, skipping the specified columns
data = pd.read_csv('Economic Data.csv', skiprows=[0,1,2,25,26,27,28,29], usecols=lambda column: column not in columns_to_exclude, index_col='Indicators')
data
```

**Observe data.head()**

```python
data.head()
```

**Import python library for plotting graph**

```python
import matplotlib.pyplot as plt
```

**Plot the Economic Indicator Analysis line plots**

```python
# Transpose the data for easier plotting
data = data.transpose()

# Plot the indicators
ax = data.plot()

# Set labels and title
plt.xlabel('Dates')
plt.ylabel('Indicator Values')
plt.title('Economic Indicator Analysis')

# Rotate x-axis labels if needed
plt.xticks(rotation=45)

# Move the legend to the side
ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))

# Display the plot
plt.show()
```

**Plot the Economic Indicator Analysis line plot separately for each indicator**

```python
# Transpose the data for easier plotting
data = data.transpose()

# Get the indicators
indicators = data.columns

# Iterate over each indicator
for indicator in indicators:
    # Create a new plot for each indicator
    plt.figure()

    # Get the data for the indicator
    indicator_data = data[indicator]

    # Plot the indicator
    plt.plot(indicator_data)

    # Set labels and title
    plt.xlabel('Dates')
    plt.ylabel('Indicator Values')
    plt.title(f'{indicator} Analysis')

    # Rotate x-axis labels if needed
    plt.xticks(rotation=45)

    # Display the legend
    plt.legend([indicator], loc='center left', bbox_to_anchor=(1, 0.5))

    # Display the plot
    plt.show()
```

```python
# Transpose the data for easier plotting
data = data.transpose()

# Get the indicators
indicators = data.columns

# Iterate over each indicator
for indicator in indicators:
    # Create a new plot for each indicator
    plt.figure()

    # Get the data for the indicator
    indicator_data = data[indicator]

    # Plot the indicator
    plt.plot(indicator_data)

    # Set labels and title
    plt.xlabel('Dates')
    plt.ylabel('Indicator Values')
    plt.title(f'{indicator} Analysis')

    # Rotate x-axis labels if needed
    plt.xticks(rotation=45)

    # Reduce the number of date values on the x-axis
    num_ticks = 10  # Change this value to adjust the number of ticks
    xtick_locations = plt.xticks()[0]
    xtick_labels = [str(date)[:10] for date in xtick_locations[::len(indicator_data) // num_ticks]]
    plt.xticks(xtick_locations[::len(indicator_data) // num_ticks], xtick_labels)

    # Display the legend
    plt.legend([indicator], loc='center left', bbox_to_anchor=(1, 0.5))

    # Display the plot
    plt.show()
```
