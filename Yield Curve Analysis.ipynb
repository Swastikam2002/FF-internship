**Import python libraries**

```python
import numpy as np
import pandas as pd
```

**Read the ED Market Data csv file**

```python
data = pd.read_csv('ED Market Data.csv', header=[0,1])
data
```

**Observe the data.head()**

```python
data.head()
```

**Calculate all the yield values**

```python
# Identify the columns containing the "Last Price" values for each commodity
price_columns = [col[0] for col in data.columns if 'Last Price' in col]

# Initialize an empty DataFrame to store the yield data
yield_data = pd.DataFrame()

# Calculate yield values for each commodity
for i, column in enumerate(price_columns):
    # Extract the commodity name from the column header
    commodity = str(column).split(' ')[0]
    
    # Get the column data
    column_data = data[column]
    
    # Calculate yield values
    last_prices = column_data.replace('#N/A N/A', 'NaN').astype(float)
    yields = (100 - last_prices) / 100
    
    # Assign meaningful column names to yield_data DataFrame
    yield_data[f"yield_{i+1}"] = yields

# Print the yield_data DataFrame
yield_data
```

**Observe the yield_data.head()**

```python
yield_data.head()
```

**Combine all the yield values in a single column**

```python
# Identify the columns containing the "Last Price" values for each commodity
price_columns = [col[0] for col in data.columns if 'Last Price' in col]

# Initialize an empty DataFrame to store the yield data
yield_data2 = pd.DataFrame()

# Calculate yield values for each commodity
for i, column in enumerate(price_columns):
    # Extract the commodity name from the column header
    commodity = str(column).split(' ')[0]
    
    # Get the column data
    column_data = data[column]
    
    # Calculate yield values
    last_prices = column_data.replace('#N/A N/A', 'NaN').astype(float)
    yields = (100 - last_prices) / 100
    
    # Assign meaningful column names to yield_data DataFrame
    yield_data2[f"yield_{i+1}"] = yields

# Create a new DataFrame with a single column named "yield"
new_yield_data = pd.DataFrame({'yield': yield_data2.values.flatten()})

# Print the new_yield_data DataFrame
new_yield_data
```

**Combine all the date values in a single column**

```python
# Get the columns containing the "Date" values
date_columns = [col[0] for col in data.columns if 'Date' in str(col)]

# Create a new DataFrame with a single column "date"
date_data = pd.DataFrame({'date': data[date_columns].values.flatten()})

# Print the date_data DataFrame
date_data
```

**Import python library for plotting graph**

```python
import matplotlib.pyplot as plt
```

**Plot the Yield Curve scatter plot**

```python
# Convert data to compatible format
x = pd.to_datetime(date_data['date']).values.flatten()
y = new_yield_data['yield'].values.flatten()

# Set the figure size
plt.figure(figsize=(12, 6))

# Plot scatter plot
plt.scatter(x, y, s=5)

# Add grid lines
plt.grid(True)

# Set labels and title
plt.xlabel('Date')
plt.ylabel('Yield')
plt.title('Yield Curve')

# Rotate x-axis labels if needed
plt.xticks(rotation=45)

# Display the plot
plt.show()
```
